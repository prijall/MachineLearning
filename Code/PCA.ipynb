```python
import pandas as pd

df=pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df['diagnosis']=breast_cancer.target
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X=df.iloc[:, :-1]
y=df.iloc[:, -1]


X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.7, shuffle=True, random_state=33)
```

```python
lr=LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

### Training model using reduced features

```python
df_corr=df.corr()['diagnosis'].abs().sort_values(ascending=False)
df_corr
```

```python
features=df_corr[df_corr>0.6].index.to_list()[1:]
features
```

```python
from sklearn.linear_model import LinearRegression

# Checking for multicolinearity:
def calculate_vif(df, features):    #vif = Variance Inflation Factor
    vif, tolerance={}, {}

    for feature in features:
        X=[f for f in features if f!= feature]
        X, y= df[X], df[feature]

    #Extraction of r-squared from the fit:
        r2=LinearRegression().fit(X, y).score(X, y)
    
    #Calculate tolerance:
    tolerance[feature]=1-r2

    #Calculate vif
    vif[feature]=1/(tolerance[feature])

    return pd.DataFrame({'VIF':vif, 'Tolerance':'tolerance'})
```

```python

calculate_vif(df, features)
```

```python
# try to reduce those feature that has high VIF until each feature 
# has VIF less than 5
features = [
    'worst concave points',
    'mean radius',
    'mean concavity',
]
calculate_vif(df,features)
```
