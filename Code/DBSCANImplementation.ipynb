```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([[1, 2], [2,2], [2, 3], [8, 7], [8, 6], [25, 80]])
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
db=DBSCAN(eps=3, min_samples=2)
db.fit(X)
```

```python
db.labels_
```

```python
from sklearn.datasets import make_circles
```

```python
#Creating a concenttric circle dataset
X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=33)
```

```python
#Apply DBSCAN to the dataset
dbscan=DBSCAN(eps=0.1, min_samples=5)
cluster=dbscan.fit_predict(X)
```

```python
#Plotting

plt.scatter(X[:, 0], X[:, 1], c=cluster, cmap='viridis', marker='o')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```
