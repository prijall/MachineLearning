```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
df=load_diabetes()
```

```python
dataset=pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names
dataset.head()
```

```python
#seperating dependent and independent features
X=dataset
y=df.target
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.70, random_state=42)
```

```python
dataset  #total data = 442
X_train  #training data = 309
X_test   #testing data =133
```

```python
#Standardizing datasset to get optimized o/p

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
mse=cross_val_score(regressor, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
```

```python
np.mean(mse)
```

```python
reg_predict= regressor.predict(X_test)
reg_predict
```

```python
sns.displot(reg_predict-y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_predict, y_test)
score
```
